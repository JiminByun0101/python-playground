```python
import folium
```

```python
map = folium.Map(location=[49.233569, -123.154106], zoom_start=10, titles='Walk In Clinics in BC')
```

```python
map
```

```python
clinics['LATITUDE'] = clinics['LATITUDE'].astype(float)
```

```python
clinics['LONGITUDE'].dtypes
```

```python
for i, clinic in clinics.iterrows():
    folium.Marker(location=[clinic['LATITUDE'], clinic['LONGITUDE']], popup = clinic['RG_NAME'], icon=folium.Icon(color='red', icon = 'star')).add_to(map)
```
